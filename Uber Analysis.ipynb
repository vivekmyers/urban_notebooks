```python
import pandas as pd
from tqdm import tqdm_notebook as tqdm
from datetime import datetime as dt
from itertools import *

# Read uber and taxis trip times for 2015
uber = pd.read_csv('resources/uber.csv', low_memory=False)
taxis = pd.read_csv('resources/taxis2015.csv', low_memory=False)

def parse(time): 
    try: 
        return dt.strptime(time, "%Y-%m-%d %H:%M:%S")
    except:
        return None
    
taxi_times = [parse(time)
                 for time in tqdm(list(taxis.tpep_pickup_datetime))
                 if parse(time)]
uber_times = [parse(time)
                 for time in tqdm(list(uber.Pickup_date))]
```

```python
import numpy as np

# Get 24D vectors for each day corresponding to hourly uber trips
vecs = np.zeros([181, 24])
day_labels = np.zeros([181], dtype=np.int)
month_labels = np.zeros([181], dtype=np.int)
for i in tqdm(uber_times):
    day_labels[i.timetuple().tm_yday - 1] = i.weekday()
    month_labels[i.timetuple().tm_yday - 1] = i.month
    vecs[i.timetuple().tm_yday - 1][i.hour] += 1

vecs = vecs / vecs.sum(axis=1)[:, np.newaxis]   
print(vecs)
```

```python
# Get 24D vectors for each day corresponding to hourly taxi trips
t_vecs = np.zeros([181, 24])
t_day_labels = np.zeros([181], dtype=np.int)
for i in tqdm(taxi_times):
    try:
        t_day_labels[i.timetuple().tm_yday - 1] = i.weekday()
        t_vecs[i.timetuple().tm_yday - 1][i.hour] += 1
    except:
        pass

t_vecs = t_vecs / t_vecs.sum(axis=1)[:, np.newaxis]   
print(t_vecs)
```

```python
import sklearn.manifold as manifold
import seaborn as sns
import calendar

# List of day names
days = list(calendar.day_abbr)

# Get 2D embedding of 24D uber vectors colored by actual day of week
em = manifold.TSNE().fit(vecs).embedding_

plt.figure()
for i in set(day_labels):
    x = np.extract(day_labels == i, em[:, 0])
    y = np.extract(day_labels == i, em[:, 1])
    plt.scatter(x=x, y=y, 
        color=sns.color_palette("hls", 7)[i], label=days[i])
    

plt.legend()
plt.axis('off')
plt.show()
```

```python
# Redo embedding including 24D taxi vectors as well
all_vecs = np.concatenate([vecs, t_vecs], axis=0)
em = manifold.TSNE().fit_transform(all_vecs)

plt.figure()

plt.scatter(x=em[181:, 0], 
            y=em[181:, 1], 
            color=sns.color_palette('hls', 2)[0], 
            label='taxis')

plt.scatter(x=em[:181, 0], 
            y=em[:181, 1], 
            color=sns.color_palette('hls', 2)[1], 
            label='uber')

plt.legend()
plt.axis('off')
plt.show()
```

```python
# Get mean 24D vectors for uber vectors and taxi vectors
u_mean = vecs.mean(axis=0)
t_mean = t_vecs.mean(axis=0)

# Classify every vector in both sets based on if it is closer to u_mean than t_mean and store in bool array
labels = np.linalg.norm(all_vecs - u_mean[np.newaxis, :], axis=1) < \
            np.linalg.norm(all_vecs - t_mean[np.newaxis, :], axis=1)
# Array of if each point is closer to "correct" mean
circles = labels ^ np.concatenate([np.repeat(True, 181), np.repeat(False, 181)])
plt.figure()

# Project all points, colored by which mean they are closest to, and circled if they are closer to the wrong mean
plt.scatter(x=em[labels, 0], 
            y=em[labels, 1], 
            color=sns.color_palette('hls', 2)[1], 
            label='uber')
plt.scatter(x=em[~labels, 0], 
            y=em[~labels, 1], 
            color=sns.color_palette('hls', 2)[0], 
            label='taxis')
plt.scatter(x=em[circles, 0],
            y=em[circles, 1],
            facecolors='none',
            edgecolors='black',
            s=50,
            linewidth=1.5,
            label='incorrect prediction')
plt.scatter(x=[], y=[], facecolors='none', edgecolors='none', 
            label=f'accuracy = {sum(~circles) * 100 / circles.size:.2f}%')

plt.legend()
plt.axis('off')
plt.show()

all_day_labels = np.concatenate([day_labels, t_day_labels])
bad_days = {k: len(list(v)) for k, v in groupby(sorted(all_day_labels[circles]))}

# Print the number of times each day is closer to the wrong mean
# High numbers should mean those days have similar distributions
for k in bad_days:
    print(f'{days[k]}\t\t{bad_days[k]}')
    
```

```python
# Redo daily vectors with 10 minute bins
u_144 = np.zeros([181, 144])
t_144 = np.zeros([181, 144])

for i in tqdm(taxi_times):
    try:
        t_144[i.timetuple().tm_yday - 1, i.hour * 6 + i.minute // 10] += 1
    except: 
        pass
    
for i in tqdm(uber_times):
    try:
        u_144[i.timetuple().tm_yday - 1, i.hour * 6 + i.minute // 10] += 1
    except:
        pass
```

```python
from functools import partial

# For each 10 minute interval, plot the taxi traffic vs the ratio of uber to taxi traffic
plt.figure()
x = np.array(list(chain(*t_144)))
y = np.array(list(chain(*u_144))) / x
# Exclude very light traffic as this skews the concavity
truncate = x > 1000
x = x[truncate & ~np.isnan(y)]
y = y[truncate & ~np.isnan(y)]
plt.scatter(x, y, s=1, alpha=0.2)
plt.xlim([-max(x) / 10, max(x) * 1.1])
plt.ylim([-max(y) / 10, max(y) * 1.1])

# Positive correlation indicates positive concavity (insufficient taxi supply)
fit = np.polyfit(x, y, 1)
r = np.corrcoef([x, y])
plt.plot(np.arange(-max(x), max(x) * 1.1, 0.1), 
         np.vectorize(partial(np.polyval, fit))(np.arange(-max(x), max(x) * 1.1, 0.1)),
         color='r',
         label=f'r = {r[0, 1]:.3f}')

plt.legend()
plt.xlabel('T')
plt.ylabel('U / T')
plt.show()
```

```python
# Create dictionary mapping start times of taxi trips to their durations in seconds
# This cell will take 1+ hours to run
taxi_duration_times = {parse(t1): (parse(t2) - parse(t1)).total_seconds()
                 for _, t1, t2 in tqdm(list(taxis[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].itertuples()))
                 if parse(t1) and parse(t2)}
```

```python
# Plot durations of taxi trips over the course of each weekday
fig, ax = plt.subplots(nrows=1, ncols=7, figsize=[20, 1.3])
print('Taxis Trip Durations')

# make array of lists of trip times for each hour of week
trip_grid = np.empty([7, 24], dtype=object)
for time, duration in taxi_duration_times.items():
    if trip_grid[time.weekday()][time.hour] is None:
        trip_grid[time.weekday()][time.hour] = []
    # remove outliers
    if duration < 86_400:    
        trip_grid[time.weekday()][time.hour].append(duration)

# reduce lists by mean
averages = np.vectorize(np.mean)(trip_grid)

for col, day in zip(ax, days):
    col.set_title(day)
    
for i, col in enumerate(ax):
    data = averages[i]
    col.plot(np.arange(len(data)), data)
    col.fill_between(np.arange(len(data)), data, alpha=0.2)
    col.axis('off')
```

```python
# Plot grid of 24D uber vectors for first half of year with labels for total calls
uber_grid = np.zeros([27, 7, 24])

for time in uber_times:
    uber_grid[int(dt.strftime(time, "%U")) 
              if time.weekday() != 6 else int(dt.strftime(time, "%U")) - 1][time.weekday()][time.hour] += 1
    
fig, ax = plt.subplots(nrows=27, ncols=7, figsize=[15, 24])

for col, day in zip(ax[0], days):
    col.set_title(day)

for i, row in enumerate(ax):
    for j, pl in enumerate(row):
        if uber_grid[i][j].sum():
            pl.plot(np.arange(24), uber_grid[i][j], label=uber_grid[i][j].sum())
            pl.fill_between(np.arange(24), uber_grid[i][j], alpha=0.2)
            pl.legend(prop={'size': 8})
        pl.axis('off')
```
